In this notebook, the appropriate data on regime type will be extracted from the "Political Regimes of the World" dataset, so that it may be merged with the overall dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Regimes_data = pd.read_csv("Political Regimes of the World Dataset, v.2.0.csv")
```

```python
Regimes_data_70=Regimes_data[Regimes_data['year'] == 1970]
Regimes_data_70.head(5)
```

```python
Regimes_70_selection = Regimes_data_70[['country','abbreviation','democracy','monarchy','regimenarrowcat']]
```
