In this notebook we will create a new variable for the V-Dem/Frasier data that notes whether America made an alliance with any of the countries in the data-set or not. A 0 will be used for no alliance, and a 1 will code for an alliance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#to be changed below
FV_data = pd.read_csv("V-Dem_Frasier_data_1970.csv")
alliance_data = pd.read_csv("US_allys_45to70.csv")
```

```python
alliance_data.head()
```

```python
alliance_entities = alliance_data['state_name2'].tolist()
FV_entities1 = FV_data['Countries'].tolist()
FV_entities2 = FV_data['Entity'].tolist()
```

```python
only_alliance=[i for i in alliance_entities if i not in FV_entities2]
print(only_alliance)
print(len(only_alliance))
```

It can thus be seen that the only discrepancy is the alliance data refers to the "German Federal Republic" which is accurate for the time period, before the fall of the Berlin Wall, but the other data refers to "Germany" for west Germany, thus a change can be made changing German federal republic to just Germany for the purposes of any merging.   

```python
alliance_data=alliance_data.replace('German Federal Republic', 'Germany')
alliance_data.head()
```

```python
alliance_data["alliance"]=1
```

```python
alliance_data.rename(columns={'state_name2': 'Entity'}, inplace=True)
```

```python
merge_try = pd.merge(FV_data,alliance_data,on='Entity',how='left')
```

```python
merge_try
```

```python
merge_try['alliance'] = merge_try['alliance'].fillna(0)
```

```python
merge_try.to_csv("V-Dem_Frasier_data_1970_alliances.csv") 
```
